```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, on=["school_name", "school_name"])
school_data_complete
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# total_schools = count by school
total_schools = len(school_data_complete["school_name"].unique())

# total_students = count by students
total_students = school_data_complete["student_name"].count()

# budget_total = total budget
budget_total = school_data_complete["budget"].unique()
budget_total = budget_total.sum()

# avg_math = average math score
avg_math = school_data_complete['math_score'].mean()

# avg_read = avgerage reading score
avg_read = school_data_complete['reading_score'].mean()

# perc_math = passing math score over 70
pass_math = school_data_complete.loc[school_data_complete["math_score"] >= 70,["math_score"]]
pass_math = pass_math.count()
perc_math = pass_math/(school_data_complete['math_score'].count())* 100

# perc_read = passing reading score over 70
pass_reading = school_data_complete.loc[school_data_complete["reading_score"] >= 70,["reading_score"]]
pass_reading = pass_reading.count()
perc_reading = (pass_reading/school_data_complete['reading_score'].count())* 100


# overall = passing both math and reading over 70
grade=[0,69,100]
mcdonalds=["Fail","Pass"]
school_data_complete["passed_math"] = pd.cut(school_data_complete["math_score"], grade, labels=mcdonalds)
school_data_complete["passed_reading"] = pd.cut(school_data_complete["reading_score"], grade, labels=mcdonalds)
school_data_complete.loc[school_data_complete['passed_math'].str.contains('Pass'), 'passed_math'] = 'Pass'
school_data_complete["both_pass"] = np.where(school_data_complete["passed_math"].str.contains("Pass") & school_data_complete["passed_reading"].str.contains("Pass"),school_data_complete["passed_math"],np.nan)
both_pass = school_data_complete["both_pass"].count()
overall = (both_pass/total_students)*100
```

```python
#summary = assign dataframe name
summary = {
    "Total Schools":[total_schools],
    "Total Students":[total_students], 
    "Total Budget":[budget_total], 
    "Average Math Score":[float(avg_math)], 
    "Average Reading Score":[float(avg_read)], 
    "% Passing Math":[float(perc_math)], 
    "% Passing Reading":[float(perc_reading)], 
    "% Overall Passing":[overall] 
}
#summary_df = summary dataframe
summary_df = pd.DataFrame(summary, columns=["Total Schools", "Total Students", "Total Budget", "Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"])
summary_df.style.hide_index()  
```
